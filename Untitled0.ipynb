```python
import pandas as pd
data={
    'name':['Alice','Bob','Claire','Tony','Jane','Purity'],
    'age':[25,30,27,21,23,34],
    'gender':['Female','Male','Female','male','female','female'],
    'marks':[90,80,'NaN',78,'NaN',54]
}
bellah=pd.DataFrame(data)
print(bellah)
c=avg=0
for i in bellah['marks']:
  if str(i).isnumeric():
    c+=1
    avg+=i
print(avg/c)
bellah=bellah.replace(to_replace='NaN',value=avg/c)
print (bellah)
#
```
